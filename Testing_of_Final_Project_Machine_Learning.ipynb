Linear Regression Machine Learning Model with 2015 Data Below

```python
# Create our features (columns except those in the original happiness dataset)
X = year_2015.drop(columns=['happiness_score','economy_gdp_per_capita',
       'family', 'health_life_expectancy', 'freedom',
       'trust_government_corruption', 'generosity', 'dystopia_residual'], axis=1)

# Create our target
y = year_2015["happiness_score"]
```

```python
# Show the resulting dataframe
X.dtypes
```

```python
X.columns
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
X
```

```python
# Perform the train, test, split of the data

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1)
X_train.shape
```

```python
# Create an instance of the linear regression model
model = LinearRegression()
```

```python
# Fit the model (training the model)
model.fit(X_train, y_train)
```

```python
# Generate predictions
y_pred = model.predict(X_test)
print(y_pred.shape)
```

```python
# Generate accuracy score
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
