```python
import pandas as pd
```

```python
columns = ["country_or_region", "score", "gdp_per_capita", "social_support", "healthy_life_expectancy", "freedom_to_make_life_choices", "generosity", "perceptions_of_corruption"]
```

```python
# create a new function
df = pd.DataFrame(columns=columns)
for Year in range (2015, 2017):
    path = f"Resources/{str(Year)}.csv"
    happiness_df = pd.read_csv(path)
    happiness_df['Year'] = Year
    df=pd.concat([df,happiness_df])
df=df.dropna(axis=1)
df
```

```python
df.to_csv("Resources/Happiness_Data_Complete.csv")
```
